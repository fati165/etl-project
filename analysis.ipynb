```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store CSV into DataFrame

```python
fishcsv = "resources/bass.csv"
#fishbass = pd.read_csv(fishcsv)
fishbasses = pd.read_csv(fishcsv, encoding="ISO-8859-1")
fishbasses.head(20)
```

### Create new data with select columns

```python
fish_data = fishbasses[['SampleID', 'NumberFishPerComp', 'Analyte', 'Unit', 'Result', 'WeightAvg g']].copy()
fish_data
```

```python
bassernozero= fish_data[fish_data["Result"]!= 0]
bassernozero
```

```python
#remove no record of weight
zeroweight= bassernozero[bassernozero["WeightAvg g"]!= 0]
zeroweight
```

```python
#remove nan sample id
bassnan= zeroweight[zeroweight["SampleID"].notnull()]
bassnan
```

```python
bassnan2= bassnan[bassnan["Result"].notnull()]
bassnan2
```

```python
Bass =  bassnan2[["SampleID", "Analyte", "Unit", "Result", "WeightAvg g"]].copy()
Bass
```

```python
#convert them to lowercase because postgress doesn't like upper
Bass.columns = Bass.columns.str.lower()
Bass
```

```python
#also remove the space and "g" from weightavg because postgres doesn't like that too
Bass.columns = ['sampleid', 'analyte', 'unit', 'result', 'weightavg']
Bass
```

# CARP DATA


```python
carpcsv = "resources/carps.csv"
#fishbass = pd.read_csv(fishcsv)
fishcarpet = pd.read_csv(carpcsv, encoding="ISO-8859-1")
fishcarpet
```

```python
carp_data = fishcarpet[['SampleID', 'Analyte', 'Unit', 'Result', 'WeightAvg g']].copy()
carp_data
```

```python
#remove zero values
carpnozero= carp_data[carp_data["Result"]!= 0]
carpnozero
```

```python
carpresult= carpnozero[carpnozero["Result"].notnull()]
carpresult
```

```python
carpid= carpresult[carpresult["SampleID"].notnull()]
carpid
```

```python
carpweight= carpid[carpid["WeightAvg g"]!= 0]
carpweight
```

```python
carpweight.columns = ['sampleid', 'analyte', 'unit', 'result', 'weightavg']
carpweight
```

### Connect to local database

```python
rds_connection_string = "admin1:12345@localhost:5433/etl"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for tables

### Use pandas to load csv converted DataFrame into database

```python
Bass.to_sql(name='bass', con=engine, if_exists='append', index=False)
```
